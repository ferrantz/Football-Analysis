### standardization

```python
from sklearn.preprocessing import StandardScaler
#from sklearn.preprocessing import scale

season =  list(df.season)
squad = list(df.Squad)
features = list(df.columns[2:])

# Separating out the features
x = df.loc[:, features].values
#print(x)

# Standardizing the features
x_std = StandardScaler().fit_transform(x)
#x_std = scale(x)
#print(x_std)


att_scaler = StandardScaler()
att_scaler.fit_transform(x)
#scaler.mean_ , scaler.var_
#scaler.inverse_transform(x_std) oppure (x_std*np.sqrt(scaler.var_))+scaler.mean_

print(np.mean(x_std, axis=0))
print(np.std(x_std, axis=0))
```

```python
import numpy as np

#mean_vec = np.mean(x_std, axis=0)
#cov_mat = (x_std - mean_vec).T.dot((x_std - mean_vec)) / (x_std.shape[0]-1)  #diviso n-1
#print('Covariance matrix \n%s' %cov_mat)

#cov_matrix = np.cov(x_std.T)
#print(cov_matrix)

cov_matrix = np.corrcoef(x_std.T) #Normalized covariance matrix
#Especially, in the field of "Finance," the correlation matrix typically used instead of the covariance matrix. 
#However, the eigendecomposition of the covariance matrix (if the input data was standardized) yields the same results as a eigendecomposition on the correlation matrix, since the correlation matrix can be understood as the normalized covariance matrix.
#print(cov_matrix)

#eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
eig_vals, eig_vecs = np.linalg.eigh(cov_matrix)
#np.linalg.eigh. This works directly for Hermitian matrices, and thus will have only real Eigen values, exactly this specific use case.

#print('Eigenvectors \n%s' %eig_vecs)
#print('\nEigenvalues \n%s' %eig_vals)

# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs[:10]:
    print(i[0])

#In principal component analysis (PCA), we get eigenvectors (unit vectors) and eigenvalues. Now, let us define loadings as
# Loadings = Eigenvectors â‹… âˆšEigenvalues.

#creating the loadings table
loadings = np.array([eig_pairs[i][1]*np.sqrt(eig_pairs[i][0]) for i in range(len(eig_vals))])
#loadings = np.array([eig_pairs[0][1]*np.sqrt(eig_pairs[0][0]),eig_pairs[1][1]*np.sqrt(eig_pairs[1][0])])

pc_list = [f'PC{i}' for i in range(1,len(features)+1)]
load_df = pd.DataFrame(data = loadings.transpose(), index=features, columns=pc_list)
load_df.iloc[:5,:5]
```

```python
import matplotlib.pyplot as plt
import matplotlib.axes

tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

pc_numbers=['PC%s' %i for i in range(1,len(eig_vals)+1)]

fig, ax = plt.subplots(figsize=(24,6))

plt.bar(pc_numbers,var_exp,width=0.8)
yticks = [f'{value} %' for value in range(0,120,20)]
#yticks = [f'{value:.0f} %' for value in var_exp]
plt.yticks(np.arange(0,120,20),yticks)

plt.plot(cum_var_exp,color='orange')
plt.scatter(pc_numbers,cum_var_exp,color='red')
plt.axhline(y=75, linestyle='--')

for i, val in enumerate(cum_var_exp):
    #if val < 99.9:
    ax.annotate(f'{val:.1f}', (pc_numbers[i],cum_var_exp[i]))

plt.show()
```

### PCA module

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

features = list(df.columns[2:])
n = len(features)
pca = PCA(n_components=n)
pca_data = pca.fit_transform(x_std)

principal_Df = pd.DataFrame(data = pca.components_.transpose(), index=features, columns=pc_list)

#principal_Df
#print(np.corrcoef(pca_data.T))

print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))

#The following code constructs the Scree plot
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, n+1)]

plt.figure(figsize=(24,4))
plt.bar(x=range(1,n+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.axes

var_exp = list(pca.explained_variance_ratio_*100)
cum_var_exp = np.cumsum(var_exp)

pc_numbers=['PC %s' %i for i in range(1,6)]#len(pca.components_)+1)]

fig, ax = plt.subplots(figsize=(10,6))
fig.suptitle('Teams Attack table PCA', fontsize=25)

plt.bar(pc_numbers,var_exp[:5],width=0.8)
yticks = [f'{value} %' for value in range(0,120,20)]
#yticks = [f'{value:.0f} %' for value in var_exp]
plt.yticks(np.arange(0,120,20),yticks)

plt.plot(cum_var_exp[:5],color='orange',linewidth=3)
plt.scatter(pc_numbers,cum_var_exp[:5],color='red',s=40)
plt.axhline(y=75, linestyle='--')

for i, val in enumerate(cum_var_exp[:5]):
    #if val < 99.9:
    ax.annotate(f'{val:.1f}%', (pc_numbers[i],cum_var_exp[i]))

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.gridspec import GridSpec

def axis_plot(axis,frst_pc,scnd_pc,pca_df,per_var): 
    dict1 = {'PC1':0,'PC2':1,'PC3':2,'PC4':3,'PC5':5,'PC6':6}
    dict2 = {'PC1':pca_df.PC1,'PC2':pca_df.PC2,'PC3':pca_df.PC3,'PC4':pca_df.PC4,'PC5':pca_df.PC5,'PC6':pca_df.PC6}
    axis.set_xlabel(f'{frst_pc} - {per_var[dict1[frst_pc]]}%')
    axis.set_ylabel(f'{scnd_pc} - {per_var[dict1[scnd_pc]]}%')
    axis.set(xlim=(-1, 1), ylim=(-1, 1))
    a_circle = plt.Circle((0, 0), 1, color='b', fill=False)
    axis.add_artist(a_circle)
    data_x = dict2[frst_pc]
    data_y = dict2[scnd_pc]
    axis.scatter(data_x, data_y)
    for sample in pca_df.index:
        axis.annotate(sample, (data_x.loc[sample], data_y.loc[sample]))
    axis.set_title(f'{frst_pc} vs. {scnd_pc}')
    axis.grid(linestyle='--')


pca_df = pd.DataFrame(pca.components_, index=features, columns=labels)

#fig, ax = plt.subplots(figsize=(12,8))
fig = plt.figure(figsize=(18,6))
plt.axis('off')
fig.suptitle('PCA Graphs',size=25)

gs=GridSpec(1,3) # 1 row, 2 columns

ax1=fig.add_subplot(gs[0,0]) # PC1, PC2
ax2=fig.add_subplot(gs[0,1]) # PC1, PC3
ax3=fig.add_subplot(gs[0,2]) # PC2, PC3

axis_plot(ax1,'PC1','PC2',pca_df,per_var)
axis_plot(ax2,'PC1','PC3',pca_df,per_var)
axis_plot(ax3,'PC2','PC3',pca_df,per_var)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.axes

tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

pc_numbers=['PC%s' %i for i in range(1,len(eig_vals)+1)]

fig, ax = plt.subplots(figsize=(24,6))

plt.bar(pc_numbers,var_exp,width=0.8)
yticks = [f'{value} %' for value in range(0,120,20)]
plt.yticks(np.arange(0,120,20),yticks)

plt.plot(cum_var_exp,color='orange')
plt.scatter(pc_numbers,cum_var_exp,color='red')
plt.axhline(y=75, linestyle='--')

for i, val in enumerate(cum_var_exp):
    ax.annotate(f'{val:.1f}', (pc_numbers[i],cum_var_exp[i]))

plt.show()
```
